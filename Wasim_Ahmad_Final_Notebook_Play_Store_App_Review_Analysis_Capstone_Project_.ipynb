# Now Fill Null Values

```python
# Fill null values in the corresponding columns
play_Store_Raw_df['Rating'] = play_Store_Raw_df['Rating'].fillna(play_Store_Raw_df['Rating'].mean())
play_Store_Raw_df['Type'].fillna('Free', inplace=True)
play_Store_Raw_df['Content Rating'].fillna('Everyone', inplace=True)
play_Store_Raw_df['Android Ver'].fillna('4.1 and up', inplace=True)
play_Store_Raw_df['Current Ver'].fillna('Varies with device', inplace=True)
```

```python
# As the null values has been filled now we can check
play_Store_Raw_df.isnull().sum()
```

* All the null values are replaced in each column

* Now we can check info

```python
play_Store_Raw_df.info()
```

* All the columns are of object data type except rating
* Lets convert these data type into int, datetime, float etc. according to their nature
* For this let's remove unwanted charechter and symbols etc. 
* These symbol and charechter can be identified from above mention First or last five rows of data set and description etc.

```python
play_Store_Raw_df['Size'] = play_Store_Raw_df['Size'].str.replace('M','e6')
play_Store_Raw_df['Size'] = play_Store_Raw_df['Size'].str.replace('k','e3')
play_Store_Raw_df['Size'] = play_Store_Raw_df['Size'].str.replace('Varies with device','0')
play_Store_Raw_df['Installs'] = play_Store_Raw_df['Installs'].str.replace('+','')
play_Store_Raw_df['Installs'] = play_Store_Raw_df['Installs'].str.replace(',','')
play_Store_Raw_df['Price'] = play_Store_Raw_df['Price'].str.replace('$','')

# Its time to have a look 
play_Store_Raw_df.head()
```

* Now change the Data type

```python
# Changing the data type according to their nature like reviews can be float type

play_Store_Raw_df['Reviews'] = play_Store_Raw_df['Reviews'].astype(float)
play_Store_Raw_df['Size'] = play_Store_Raw_df['Size'].astype(float)
play_Store_Raw_df['Installs'] = play_Store_Raw_df['Installs'].astype(int)
play_Store_Raw_df['Price'] = play_Store_Raw_df['Price'].astype(float)
```
